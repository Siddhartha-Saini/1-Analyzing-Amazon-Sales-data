```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
#set seaborn style 
sns.set(style='whitegrid')
```

```python
file_path = 'Amazon Sales data.csv'
df = pd.read_csv('Amazon Sales data.csv')
```

```python
print(df.head())
```

```python
print(df['Order Date'].dtype) 
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
print(df['Order Date'].dtype) 
```

```python
df['year'] = df['Order Date'].dt.year
df['month'] = df['Order Date'].dt.month
df['year_month'] = df['Order Date'].dt.to_period('M')
```

```python
print(df)
```

```python
df
```

```python
df.rename(columns={'Total Revenue': 'Sales'}, inplace=True)
```

```python
print(df.head())
```

```python
month_wise_sales = df.groupby('month')['Sales'].sum()
year_wise_sales = df.groupby('year')['Sales'].sum()
yearly_month_wise_sales = df.groupby('year_month')['Sales'].sum()
```

```python
print("Month-wise Sales:\n", month_wise_sales)
print("Year-wise Sales:\n", year_wise_sales)
print("Yearly Month-wise Sales:\n", yearly_month_wise_sales)
```

```python
df
```

```python
total_sales = df['Sales'].sum()
average_order_values = df['Sales'].mean()
number_of_order = df.shape[0]
```

```python
print("Total Sales:", total_sales)
print("Average Order Values:", average_order_values)
print("Number of Orders:", number_of_order)
```

```python
numeric_df = df.select_dtypes(include=['number'])
```

```python
correlation_matrix = numeric_df.corr()
```

```python
print("correlation_matrix:\n", correlation_matrix)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Month-wise Sales Plot
plt.figure(figsize=(10, 6))
month_wise_sales.plot(kind='bar')
plt.title('Month-wise Sales')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.show()
```

```python
# Year-wise Sales Plot
plt.figure(figsize=(10, 6))
year_wise_sales.plot(kind='bar')
plt.title('Year-wise Sales')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.show()
```

```python
# Yearly Month-wise Sales Plot
plt.figure(figsize=(14, 7))
yearly_month_wise_sales.plot()
plt.title('Yearly Month-wise Sales')
plt.xlabel('Year-Month')
plt.ylabel('Sales')
plt.show()
```

```python
# Heatmap for correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
